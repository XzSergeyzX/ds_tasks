```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
# Look at the first 5 rows
print("--- first 5 rows ---")
display(df.head())

#  Data info / checking missing values
print("\n--- Data info / checking missing values ---")
df.info()

# Describing data
print("\n--- Columns description ---")
display(df.describe())
```

### Dataset overview

The dataset contains **90,000 rows and 54 columns**.

All features are numeric, consisting of:
- 43 float features
- 11 integer features

No missing values are present in the dataset, so no imputation step is required.

The total memory footprint of the dataset is approximately **37 MB**, which allows efficient training with standard machine learning models.

Since all features are numeric, no categorical encoding is required.
This simplifies the modeling pipeline.

```python
# Target column
target_col = df.columns[-1]

# Target distribution
plt.figure(figsize=(10, 6))
sns.histplot(df[target_col], kde=True)
plt.title('Distribution of Target Variable')
plt.show()

# Corelation with target
correlations = df.corr()[target_col].drop(target_col).sort_values(ascending=False)

print("Top 10 correlated features with target:")
print(correlations.head(10))
```

### Target distribution

The target variable appears to be approximately **uniformly distributed** across the range from 0 to 100.

There is no visible skewness or strong concentration of values in a specific region of the range.  
This suggests that the regression problem is well-balanced across the target space.

Since the target distribution is relatively flat, evaluation metrics such as **RMSE** should provide a stable estimate of model performance across the dataset.

### Feature correlation with target

Next, we compute the correlation between each feature and the target variable in order to inspect potential **linear relationships**.

If correlations are generally weak, purely linear models may struggle to capture the underlying relationships between features and the target.

In such cases, **tree-based models** are often a good choice because they can model **non-linear interactions and feature combinations** without requiring extensive feature engineering.

Based on this observation, a **Random Forest Regressor** is selected as the baseline model.
